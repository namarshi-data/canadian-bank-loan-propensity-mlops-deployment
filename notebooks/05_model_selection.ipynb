# 05 Model Selection

This notebook documents the final model selection decision for the bank loan propensity prediction project. The focus is on translating model metrics into a business-ready decision for targeted banking campaigns.

## Selection Criteria

For a Canadian banking analytics portfolio project, the final model should be evaluated on both statistical performance and business usefulness. The selected model should balance:

- **ROC-AUC** for ranking quality
- **Precision** to avoid unnecessary campaign spend
- **Recall** to capture likely borrowers
- **F1 Score** to balance precision and recall
- **False Negatives** because missed high-propensity customers represent lost revenue opportunities
- **False Positives** because irrelevant outreach increases campaign cost and customer fatigue

## Final Decision

The tuned **Hist Gradient Boosting** model is selected as the final production model because it provides the best balance across ranking quality, precision, recall, and operational campaign cost. It identifies most likely borrowers while keeping unnecessary outreach low.

## Business Interpretation

Out of 96 actual borrowers in the test set, the selected model correctly identified 92 borrowers and missed only 4. It also generated only 2 false positives, which keeps marketing waste low. This is a strong fit for a targeted banking campaign where both conversion lift and customer experience matter.

## Alternative Business Model

If the bank decides that missing even one potential borrower is more costly than sending extra campaign offers, the Random Forest + SMOTE + Undersampling model can be used as a recall-maximizing alternative. This kind of trade-off is realistic in banking because the best model depends on the campaign budget, customer-contact policy, and risk appetite.
