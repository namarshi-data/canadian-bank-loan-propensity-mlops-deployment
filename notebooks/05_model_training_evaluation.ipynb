## Business Interpretation

For a loan marketing campaign:

- **False positives** represent customers targeted unnecessarily.
- **False negatives** represent likely borrowers missed by the campaign.
- **Precision** controls campaign efficiency.
- **Recall** controls missed opportunity.
- **F1 score** balances both.

```python
business_columns = [
    "Model",
    "Precision",
    "Recall",
    "F1 Score",
    "ROC-AUC",
    "False Positives",
    "False Negatives",
]
display(comparison[business_columns])
```
